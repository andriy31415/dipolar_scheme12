```python
%matplotlib widget
# %matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from ipywidgets import interact
import ipywidgets as widgets
from math import pi
from matplotlib import cm
import import_ipynb
from scheme_3D import Dipolar_matrix


# from matplotlib.colors import LightSource
# ls = LightSource(azdeg=315, altdeg=45)


# This is plot which demonstate surface of how density gradient vector shifts, with all possible directions of magetic fields
def plot_3D_ellipsoid2(k=1.01, e_dd=0.7,theta_A_deg=76):
    ax.clear()
    phi_A_deg=0
    phi_A = np.radians(phi_A_deg)  
    theta_A = np.radians(theta_A_deg)

    A_vec=-np.array([np.cos(phi_A)*np.sin(theta_A),np.sin(phi_A)*np.sin(theta_A),np.cos(theta_A)])

    N_p=128;
    phi_lin=np.linspace(0,2*pi,N_p)
    theta_lin=np.linspace(0,pi,N_p)
    X_surf=np.zeros((N_p,N_p))
    Y_surf=np.zeros((N_p,N_p))
    Z_surf=np.zeros((N_p,N_p))
    for i in range(len(phi_lin)):
        for j in range(len(theta_lin)):
            C1=Dipolar_matrix(k, e_dd, theta_lin[i], phi_lin[j])
            A0=-C1@A_vec+A_vec
            X_surf[i,j]=A0[0]
            Y_surf[i,j]=A0[1]
            Z_surf[i,j]=A0[2]
            
    # rgb = ls.shade(
    # Z_surf,
    # cmap=plt.cm.gray,   # single-tone base
    # vert_exag=1,
    # blend_mode="soft"
    # )
    # ax.plot_surface(X_surf,Y_surf,Z_surf, alpha=0.7, facecolors=rgb, linewidth=0, antialiased=True)
    ax.plot_surface(X_surf,Y_surf,Z_surf, alpha=0.5, cmap=cm.hot)
    ax.scatter(0,0,0,color='blue',s=8);
    ax.set_aspect('equal')
    # ax.set_xlim(-1.5,1.5);
    # ax.set_ylim(-1.5,1.5);
    # ax.set_zlim(-1.5,1.5);
    plt.show();
    # plt.close(f);  
    f.canvas.draw_idle()
    
f = plt.figure(figsize=(8, 5));
ax=plt.axes(projection='3d');

interact(plot_3D_ellipsoid2,
         k=(1,8,0.01),
         e_dd=(0,1,0.1),
         theta_A_deg=(0, 180, 2),
         );
```
