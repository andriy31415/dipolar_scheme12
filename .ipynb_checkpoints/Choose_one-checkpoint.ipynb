```python
import import_ipynb
import Scheme_circle as el_XY
import Scheme_ellipse as el_XZ
import numpy as np
import matplotlib.pyplot as plt
import ipywidgets as widgets
# from ipywidgets import interact
from IPython.display import display, clear_output, Markdown

# Dropdown menu
dropdown = widgets.Dropdown(
    options={"X-Y plane": "XY", "X-Z plane": "XZ"},
    description='Choose view:'
)
display(Markdown(r"""Here is diagram on how acceleration affects density distribution for prolate/oblate ellipsoid of the form: $x^2+y^2+k^2 z^2 = C$
"""))


k_slider=widgets.FloatSlider(min=0.35, max=3, step=0.05, description="k")
edd_slider=widgets.FloatSlider(min=0, max=1, step=0.1, description=r"$\varepsilon_{dd}$")
phi_slider=widgets.IntSlider(min=0, max=90, step=1, description=r"$\phi$")
theta_slider=widgets.IntSlider(min=0, max=90, step=1, description=r"$\theta$")

plot_output = widgets.Output()
text_output = widgets.Output()
controls_box = widgets.VBox()

# Updates plot function
def update_plot():
        with plot_output:
            clear_output(wait=True)  # Clear previous plot
            if dropdown.value == 'XY':
               el_XY.plot_ellipse_XY(k_slider.value, edd_slider.value, phi_slider.value, theta_slider.value)
            elif dropdown.value == 'XZ':
               el_XZ.plot_ellipse_XZ(k_slider.value, edd_slider.value, phi_slider.value, theta_slider.value)

def update_text():
    shared_text=r"""
Left side: display plane view of dipole distribution inside ellipsoid (the total volume assumed to be fixed)
Right side: display how density gradient induced by acceleration is deformed due to dipolar interaction.

Angle $\theta$ regulates direction of the acceleration (red vector), while $\phi$ direction of the magnetic field (purple vector)
The green vector shows the total direction of density gradient. Due to dipolar interaction it can change as by direction, as well by amplitude. If the dipolar interaction is absent, the density gradient is simply opposite to acceleration.
Cyan dashed line depict axis of zero deflection: if magnetic field is directed along them, the density gradient is just scaled, and not deflected. Such lines do not depend on $\varepsilon_{{dd}}$.  
"""

    # md_text = ""
    if dropdown.value == "XY":
        md_text = r"""
***
XY-view
"""
    elif dropdown.value == "XZ":
        md_text = r"""
***
XZ-view
"""
    full_text=md_text+shared_text
    with text_output:
        clear_output(wait=True)
        display(Markdown(full_text))
        
def update_controls(change=None):
    if dropdown.value == "XY":
        controls_box.children = [k_slider, edd_slider, phi_slider, theta_slider]
    elif dropdown.value == "XZ":
        controls_box.children = [k_slider, edd_slider, phi_slider, theta_slider]
    update_plot()
    update_text()

dropdown.observe(update_controls, names="value")
for slider in [k_slider, edd_slider, phi_slider, theta_slider]:
    slider.observe(lambda change: [update_plot(), update_text()], names="value")

update_controls()  # set up initial controls
display(widgets.VBox([dropdown, controls_box, plot_output, text_output]))
```
